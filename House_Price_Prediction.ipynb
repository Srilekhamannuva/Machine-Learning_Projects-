## Corelationship between categorical Columns and continuous target variable using one way anova

```python
from scipy.stats import f_oneway
```

```python
df_cat['sp']=df_num['SalePrice']
```

```python
groups = [df_cat['sp'][df_cat['BsmtExposure'] == category] for category in df_cat['BsmtExposure'].unique()]

# Perform ANOVA
f_stat, p_value = f_oneway(*groups)
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

```python
c_list=[]
nc_list=[]
for c1 in list(df_cat.columns):
    if c1 == 'sp':
        continue
    else:
        groups = [df_cat['sp'][df_cat[c1] == category] for category in df_cat[c1].unique()]
        f_stat, p_value = f_oneway(*groups)
        print(f"column : {c1}, F-statistic: {f_stat}, P-value: {p_value}")
        if p_value < 0.05:
            c_list.append(c1)
        else:
            nc_list.append(c1)
```

## IQR TREATMENT

```python
def remove_outliers(df,columns):
    for col in columns:
        q1=df[col].quantile(0.25) #first quartile 25th percentile
        q3=df[col].quantile(0.75) #third quartile 75th percentile
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3*1.5*iqr
        df=df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
    return df
```

```python
columns_to_check = ['LotFrontage',
 'OverallQual',
 'YearBuilt',
 'YearRemodAdd',
 'MasVnrArea',
 'BsmtFinSF1',
 'TotalBsmtSF',
 '1stFlrSF',
 '2ndFlrSF',
 'GrLivArea',
 'FullBath',
 'TotRmsAbvGrd',
 'Fireplaces',
 'GarageYrBlt',
 'GarageCars',
 'GarageArea',
 'WoodDeckSF',
 'OpenPorchSF',
 'SalePrice'] 
df_no_outliers = remove_outliers(df_num, columns_to_check)

print("Original DataFrame:")
print(df_num.shape)
print("\nDataFrame after Outlier Treatment:")
print(df_no_outliers.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

# Define Linear Regression model
lin_reg = LinearRegression()

# Define the valid parameter grid for Linear Regression
param_grid = {
    'fit_intercept': [True, False],  # Whether to calculate the intercept
    'copy_X': [True, False],  # Whether to copy input X or overwrite it
    'positive': [True, False]  # Whether to enforce positive coefficients (added in sklearn 0.24+)
}

# Instantiate GridSearchCV
gdf = GridSearchCV(estimator=lin_reg, param_grid=param_grid, scoring='r2', cv=5)

# Fit to the training data
gdf.fit(x_train, y_train)

# Get the best parameters and best score
print("Best Parameters:", gdf.best_params_)
print("Best Score:", gdf.best_score_)
print("Best Score:", gdf.best_score_)
```
